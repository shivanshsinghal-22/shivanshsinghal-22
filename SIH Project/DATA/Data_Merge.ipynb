```python
wrm_2_data = pd.read_excel('/content/WRM_2_DATA.xlsx')
```

```python
wrm_2_data.columns
```

```python
weight_data = pd.read_excel('/content/WEIGHT_DATA.xlsx')
```

```python
weight_data.columns
```

```python
weight_data = weight_data[['castNo', 'coilNo','rawDate']]
```

```python
weight_data = weight_data[((weight_data['castNo'] // 100000) == 7) | ((weight_data['castNo'] // 100000) == 8)]
```

```python
weight_data.dropna()
```

```python
import pandas as pd

# Convert columns to datetime
weight_data['rawDate'] = pd.to_datetime(weight_data['rawDate'])
wrm_2_data['Timestamp_UTC'] = pd.to_datetime(wrm_2_data['timestamp_utc'])

# Sort WRM_2_DATA by Timestamp_UTC to simplify finding the closest match
wrm_2_data = wrm_2_data.sort_values('Timestamp_UTC')
weight_data = weight_data.sort_values('rawDate', ascending=False)

# Initialize a list to store the matching rows
merged_rows = []

# Iterate over rows in weight_data
for _, row in weight_data.iterrows():
    raw_date = row['rawDate']

    # Find rows in wrm_2_data where Timestamp_UTC is smaller or equal to rawDate - 20 minutes
    filtered_rows = wrm_2_data[wrm_2_data['Timestamp_UTC'] <= raw_date - pd.Timedelta(minutes=20)]

    if not filtered_rows.empty:
        # Select the row with the maximum Timestamp_UTC (just smaller or equal)
        closest_row = filtered_rows.iloc[[-1]]  # The last row in the filtered DataFrame

        # Combine the weight_data row with the closest matching WRM_2_DATA row
        merged_row = row.to_dict()
        merged_row.update(closest_row.iloc[0].to_dict())
        merged_rows.append(merged_row)

        # Drop the matched row from wrm_2_data to avoid using it again
        wrm_2_data = wrm_2_data.drop(closest_row.index)

# Convert the merged rows into a DataFrame
merged_df = pd.DataFrame(merged_rows)

# Optional: Save the merged DataFrame to a file
merged_df.to_excel('/content/MERGED_OUTPUT.xlsx', index=False)

print("Merging complete. Data saved to MERGED_OUTPUT.xlsx")
```
